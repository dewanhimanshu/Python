```python
import matplotlib.pyplot as plt 
# Assuming five subjects which are OOPS , CSA, DM , TC and  MTCS.
class Student:
    def __init__(self,name,marks):
        #Construtor  Args: name,marks
        self.name = name
        self.marks = marks

    def get_marks(self):
        #Return the marks of the student
        return self.marks

    def get_name(self):
        #Return the name of the studnet
        return self.name







```

```python
def main():
    n=int(input("Enter total number of students : ")) 
    student_list=[]
    subjects=['OOPS' , 'CSA' , 'DM' , 'TC' ,'MTCS']
    for i in range(n): 
        print("\nEnter details of student ",i+1," : ") 
        name=input("\tEnter Student Name : ")
        marks=[] 
        for j in subjects : 
            print("\tEnter marks in subject ",j," : ",end='')
            marks.append(int(input())) 
        student_obj = Student(name,marks) 
        student_list.append(student_obj) 

   
    while True:
        print("***** Menu Driven Program *****")
        print("1-Individual student records in form of bar graph.")
        print("2-Individual student records in form of line graph.")
        print("3-Individual student pie chart analysis.")
        print("4-Subject wise Analysis.")
        print("5.Exit")
        c=int(input("\nEnter your choice : "))
        
        if c==1:
            for i in student_list:
                j=1
                plt.subplot(n,1,j)
                j=j+1
                x = subjects
                y = i.get_marks()
                plt.bar(x, y)
                plt.title(i.name)
                plt.show()
        elif c==2:
            for i in student_list:
                j=1
                plt.subplot(n,1,j)
                j=j+1
                x = subjects
                y = i.get_marks()
                plt.plot(x, y,marker = 'o')
                plt.title(i.name)
                plt.show()
        elif c==3:
            for i in student_list:
                j=1
                plt.subplot(n,1,j)
                j=j+1
                x = subjects
                y = i.get_marks()
                plt.pie(y,labels = x)
                plt.title(i.name)
                plt.show()
        elif c==4:
            oops=[]           
            csa=[]
            dm=[]
            tc=[]
            mtcsa=[]
            names=[]
            
            for i in student_list:
                marks_list=i.get_marks()
                names.append(i.get_name())
                oops.append(marks_list[0])
                csa.append(marks_list[1])
                dm.append(marks_list[2])
                tc.append(marks_list[3])
                mtcsa.append(marks_list[4])

            
             
            x = names
            y = oops
            plt.bar(x,y)
            plt.xlabel("Students Names")
            plt.ylabel("Marks in oops")
            plt.show()

            y = csa
            plt.bar(x,y)
            plt.xlabel("Students Names")
            plt.ylabel("marks in csa")
            plt.show()
            
            y = dm
            plt.bar(x,y)
            plt.xlabel("Students Names")
            plt.ylabel("marks in dm")
            plt.show()
        
            y = tc
            plt.bar(x,y)
            plt.xlabel("Students Names")
            plt.ylabel("marks in tc")
            plt.show()
           
            y = mtcsa
            plt.bar(x,y)
            plt.xlabel("Students Names")
            plt.ylabel("marks in mtcsa")
            plt.show()
                
        else:
            break
            


if __name__ == "__main__":
    main()

```
